```python
import pandas as pd

ventas = [
    {"vendedor": "Ana", "mes": "enero", "monto": 15000, "region": "norte", "producto": "laptop"},
    {"vendedor": "Luis", "mes": "enero", "monto": 8000, "region": "sur", "producto": "mouse"},
    {"vendedor": "Ana", "mes": "febrero", "monto": 20000, "region": "norte", "producto": "laptop"},
    {"vendedor": "Eva", "mes": "enero", "monto": 12000, "region": "norte", "producto": "teclado"},
    {"vendedor": "Luis", "mes": "febrero", "monto": 9500, "region": "sur", "producto": "monitor"},
    {"vendedor": "Ana", "mes": "marzo", "monto": 18000, "region": "norte", "producto": "laptop"},
    {"vendedor": "Eva", "mes": "febrero", "monto": 15000, "region": "norte", "producto": "monitor"},
    {"vendedor": "Luis", "mes": "marzo", "monto": 11000, "region": "sur", "producto": "laptop"},
    {"vendedor": "Eva", "mes": "marzo", "monto": 9000, "region": "norte", "producto": "mouse"},
    {"vendedor": "Ana", "mes": "abril", "monto": 22000, "region": "norte", "producto": "laptop"},
]

df = pd.DataFrame(ventas)
df
```

```python
print(df.head())      # primeras 5 filas
print(df.shape)       # (filas, columnas)
print(df.columns)     # nombres de columnas
print(df.dtypes)      # tipo de dato de cada columna
print(df.info())      # resumen completo
```

```python
# Una columna - retorna una Serie
print(df['vendedor'])

# Multiples columnas - retorna un DataFrame
print(df[['vendedor', 'monto']])

# Filtrar filas - boolean indexing, igual que numpy
print(df[df['monto'] > 13000])

# Filtrar con multiples condiciones
print(df[(df['region'] == 'norte') & (df['monto'] > 13000)])
```

```python
# Total por vendedor - igual que GROUP BY + SUM
print(df.groupby('vendedor')['monto'].sum())

print("\n")
# Multiples agregaciones
print(df.groupby('vendedor')['monto'].agg(['sum', 'mean', 'count']))

print("\n")
# Agrupar por dos columnas
print(df.groupby(['region', 'vendedor'])['monto'].sum())
```

```python
# SQL : SELECT vendedor, SUM(monto) FROM ventas GROUP BY vendedor
df.groupby('vendedor')['monto'].sum()

# SQL: SELECT vendedor, SUM(monto), AVG(mono), COUNT(*) FROM ventas GROUP BY vendedor
df.groupby('vendedor')['monto'].agg(['sum', 'mean', 'count'])

# SQL: SELECT region, vendedor, SUM(monto) FROM ventas GROUP BY region, vendedor
df.groupby(['region', 'vendedor'])['monto'].sum()
```

```python
# 1. Total de ventas por región (ya lo hiciste en SQL)
df.groupby('region')['monto'].sum()
```

```python
# 2. Promedio de ventas por vendedor
df.groupby('vendedor')['monto'].mean()
```

```python
# 3. Cuántas ventas hizo cada vendedor (count)
df.groupby('vendedor')['monto'].count()
```

```python
# 4. Top 3 productos por monto total — usa groupby + sort_values + head
df.groupby('producto')['monto'].sum().sort_values(ascending=False).head(3)
```

```python
import numpy as np

ventas_sucias = pd.DataFrame({
    "vendedor": ["Ana", "Luis", "Ana", "Eva", None, "Luis", "Eva"],
    "monto": [15000, 8000, 20000, np.nan, 12000, 9500, 9000],
    "region": ["norte", "sur", "norte", "norte", "norte", "sur", "norte"],
    "producto": ["laptop", "mouse", "laptop", "teclado", "monitor", "monitor", "mouse"]
})

ventas_sucias_dup = pd.concat([ventas_sucias, ventas_sucias.iloc[[0]]])

print(ventas_sucias_dup)
```

```python
# Detectar nulos
print(ventas_sucias_dup.isnull())
print(ventas_sucias_dup.isnull().sum())

# Detectar duplicados
print(ventas_sucias_dup.duplicated())
```

```python
# Eliminar duplicados
df_limpio = ventas_sucias_dup.drop_duplicates()
print("Sin duplicados:")
print(df_limpio)

# Eliminar filas con cualquier nulo
df_sin_nulos = df_limpio.dropna()
print("\nSin nulos:")
print(df_sin_nulos)

# Alternativa: rellenar nulos en lugar de eliminar
df_relleno = df_limpio.copy()
df_relleno['monto'] = df_relleno['monto'].fillna(df_relleno['monto'].mean())
df_relleno['vendedor'] = df_relleno['vendedor'].fillna('Desconocido')
print("\nCon valores rellenados:")
print(df_relleno)
```

```python
fechas_df = pd.DataFrame({
    "fecha": ["2024-01-15", "2024-02-20", "2024-03-10", "2024-04-05"],
    "monto": [15000, 20000, 18000, 22000]
})

# Convertir a datetime
fechas_df['fecha'] = pd.to_datetime(fechas_df['fecha'])
print(fechas_df.dtypes)

# Extraer componentes
fechas_df['año'] = fechas_df['fecha'].dt.year
fechas_df['mes'] = fechas_df['fecha'].dt.month
fechas_df['dia_semana'] = fechas_df['fecha'].dt.day_name()

print(fechas_df)
```

```python
fechas = pd.date_range(start="2024-01-01", end="2024-02-29", freq="D")
ventas_diarias = pd.DataFrame({
    "fecha": fechas,
    "cantidad": np.random.randint(5, 30, size=len(fechas))
})

print(ventas_diarias.head())
print(ventas_diarias.shape)
```

```python
ventas_diarias = ventas_diarias.set_index('fecha')
print(ventas_diarias.head())

# Ahora puedes hacer resample - agrupar por periodos de tiempo
semanal = ventas_diarias.resample('W').sum()
print("\nVentas semanales:")
print(semanal)
```

```python
mensual = ventas_diarias.resample('ME').sum()
print(mensual)
```

```python
productos = pd.DataFrame({
    "producto_id": [1, 2, 3],
    "nombre": ["laptop", "mouse", "monitor"],
    "categoria": ["electronica", "accesorios", "electronica"]
})

ventas = pd.DataFrame({
    "producto_id": [1, 2, 1, 3, 2],
    "cantidad": [5, 10, 3, 2, 8],
    "fecha": ["2024-01-01", "2024-01-01", "2024-01-02", "2024-01-02", "2024-01-03"]
})

df_completo = ventas.merge(productos, on="producto_id")
print(df_completo)
```

```python
# LEFT JOIN — mantiene todas las filas de ventas aunque no tengan match
df_left = ventas.merge(productos, on="producto_id", how="left")
print(df_left)
```

```python
ventas_con_huerfano = pd.DataFrame({
    "producto_id": [1, 2, 1, 3, 2, 99],  # 99 no existe en productos
    "cantidad": [5, 10, 3, 2, 8, 15],
    "fecha": ["2024-01-01", "2024-01-01", "2024-01-02", "2024-01-02", "2024-01-03", "2024-01-04"]
})

# INNER - solo filas con match en ambas tablas
print("INNER:")
print(ventas_con_huerfano.merge(productos, on="producto_id", how="inner"))

# LEFT - todas las filas de ventas, aunque no tengan match
print("LEFT:")
print(ventas_con_huerfano.merge(productos, on="producto_id", how="left"))
```
