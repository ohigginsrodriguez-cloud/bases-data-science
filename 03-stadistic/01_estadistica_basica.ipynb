```python
import numpy as np
from scipy.stats import mode
```

```python
datos = [15000, 8000, 20000, 12000, 9500, 18000, 15000, 11000, 9000, 22000]

print("Media:", np.mean(datos))
print("Mediana:", np.median(datos))
print("Moda:", mode(datos).mode)
print("Desviación estándar:", np.std(datos))
print("Varianza:", np.var(datos))
print("Mínimo:", np.min(datos))
print("Máximo:", np.max(datos))
```

```python
print("Diferencia media-mediana:", np.mean(datos) - np.median(datos))
print("Rango: ", np.max(datos) - np.min(datos))
```

```python
print("Percentil 25:", np.percentile(datos, 25))
print("Percentil 50:", np.percentile(datos, 50))
print("Percentil 75:", np.percentile(datos, 75))
print("IQR:", np.percentile(datos, 75) - np.percentile(datos, 25))
```

```python
q25 = np.percentile(datos, 25)
q75 = np.percentile(datos, 75)
iqr = q75 - q25

limite_bajo = q25 - 1.5 * iqr
limite_alto = q75 + 1.5 * iqr

print("Límite bajo:", limite_bajo)
print("Límite alto:", limite_alto)
print("Outliers:", [x for x in datos if x < limite_bajo or x > limite_alto])
```
