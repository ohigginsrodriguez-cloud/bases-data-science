```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
ventas = pd.DataFrame({
    "mes": ["enero", "febrero", "marzo", "abril", "mayo", "junio"],
    "monto": [15000, 20000, 18000, 22000, 19000, 25000]
})

# Grafica de linea - tendencia en el tiempo
plt.figure(figsize=(10, 4))
plt.plot(ventas['mes'], ventas['monto'], marker='o')
plt.title("Ventas mensuales")
plt.xlabel("Mes")
plt.ylabel("Monto")
plt.grid(True)
plt.show()
```

```python
ventas_producto = pd.DataFrame({
    "producto": ["laptop", "mouse", "monitor", "teclado"],
    "total": [86000, 17000, 24500, 12000]
})

plt.figure(figsize=(8, 4))
plt.bar(ventas_producto['producto'], ventas_producto['total'], color='steelblue')
plt.title("Ventas totales por producto")
plt.xlabel("Producto")
plt.ylabel("Total")
plt.show()
```

```python
datos = pd.DataFrame({
    "vendedor": ["Ana", "Luis", "Eva", "Ana", "Luis", "Eva", "Ana"],
    "monto": [15000, 8000, 12000, 20000, 9500, 15000, 18000]
})

plt.figure(figsize=(8, 4))
sns.boxplot(data=datos, x='vendedor', y='monto')
plt.title("Distribución de ventas por vendedor")
plt.show()
```

```python
# Histograma - distribucion de una variable
datos =  np.random.normal(loc=15000, scale=3000, size=200)

plt.figure(figsize=(8, 4))
plt.hist(datos, bins=20, edgecolor='black', color='steelblue')
plt.title("Distribucion de montos de venta")
plt.xlabel("Monto")
plt.ylabel("Frecuencia")
plt.show()
```

```python
df_corr = pd.DataFrame({
    "monto": [15000, 8000, 20000, 12000, 9500, 18000, 15000, 11000, 9000, 22000],
    "cantidad": [3, 1, 4, 2, 1, 3, 3, 2, 1, 4],
    "dias_credito": [30, 15, 30, 30, 15, 30, 30, 15, 15, 30]
})

correlaciones = df_corr.corr()
print(correlaciones)

plt.figure(figsize=(6, 5))
sns.heatmap(correlaciones, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Matriz de correlacion")
plt.show()
```
