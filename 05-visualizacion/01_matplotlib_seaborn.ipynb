```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
ventas = pd.DataFrame({
    "mes": ["enero", "febrero", "marzo", "abril", "mayo", "junio"],
    "monto": [15000, 20000, 18000, 22000, 19000, 25000]
})

# Grafica de linea - tendencia en el tiempo
plt.figure(figsize=(10, 4))
plt.plot(ventas['mes'], ventas['monto'], marker='o')
plt.title("Ventas mensuales")
plt.xlabel("Mes")
plt.ylabel("Monto")
plt.grid(True)
plt.show()
```

```python
ventas_producto = pd.DataFrame({
    "producto": ["laptop", "mouse", "monitor", "teclado"],
    "total": [86000, 17000, 24500, 12000]
})

plt.figure(figsize=(8, 4))
plt.bar(ventas_producto['producto'], ventas_producto['total'], color='steelblue')
plt.title("Ventas totales por producto")
plt.xlabel("Producto")
plt.ylabel("Total")
plt.show()
```

```python
datos = pd.DataFrame({
    "vendedor": ["Ana", "Luis", "Eva", "Ana", "Luis", "Eva", "Ana"],
    "monto": [15000, 8000, 12000, 20000, 9500, 15000, 18000]
})

plt.figure(figsize=(8, 4))
sns.boxplot(data=datos, x='vendedor', y='monto')
plt.title("Distribución de ventas por vendedor")
plt.show()
```
