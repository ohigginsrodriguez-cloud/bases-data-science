```python
def decorador(funcion):
    def wrapper(nombre):
        print("Antes de ejecutar")
        resultado = funcion(nombre)
        print("Despues de ejecutar")
        return resultado
    return wrapper

@decorador
def saludar(nombre):
    return f"Hola {nombre}"

print(saludar("Ana"))
```

```python
import time

def timer(funcion):
    def wrapper(*args, **kwargs):
        inicio = time.time()
        resultado = funcion(*args, **kwargs)
        fin = time.time()
        print(f"{funcion.__name__} tardó {fin - inicio:.2f} segundos")
        return resultado
    return wrapper

@timer
def entrenar_modelo(datos):
    time.sleep(2)  # simulando entrenamiento
    return "modelo entrenado"

entrenar_modelo([1, 2, 3])
```

```python
# sin context manager (forma mala)
archivo = open("prueba.txt", "w")
archivo.write("Hola mundo")
archivo.close() # si algo falla antes, el archivo nunca se cierra

# con context manager (forma correcta)
with open("prueba.txt", "w") as archivo:
    archivo.write("Hola mundo")
# se cierra solo, pase lo que pase
```

```python
# Datos para los ejercicios
ventas = [
    {"vendedor": "Ana", "mes": "enero", "monto": 15000, "region": "norte"},
    {"vendedor": "Luis", "mes": "enero", "monto": 8000, "region": "sur"},
    {"vendedor": "Ana", "mes": "febrero", "monto": 20000, "region": "norte"},
    {"vendedor": "Eva", "mes": "enero", "monto": 12000, "region": "norte"},
    {"vendedor": "Luis", "mes": "febrero", "monto": 9500, "region": "sur"},
]
```

```python
# 1. Crea un decorador llamado `log` que imprima el nombre 
# de la función y cuánto tardó en ejecutarse.
# Úsalo en una función que calcule el total de ventas.
import time
def log(funcion):
    def wrapper(*args, **kwargs):
        start = time.time()
        execution = funcion(*args, **kwargs)
        end = time.time()

        print(f"{funcion.__name__} tardo: {end - start:.2f} en ejecutarse")

        return execution
    return wrapper

@log
def total_ventas(ventas: list):
    total = 0
    for v in ventas:
        total += v['monto']

    return f"El total de ventas es: {total}"

total_ventas(ventas)
```

```python
# 2. Crea un generador que reciba la lista de ventas 
# y entregue una venta a la vez, solo si el monto es mayor a 10000.
def generador(ventas: list):
    for v in ventas:
        if v['monto'] > 10000:
            yield v

for g in generador(ventas):
    print(g)
```

```python
# 3. Usando el generador del ejercicio 2, 
# recórrelo con un for e imprime vendedor y monto.
def vendedor_monto(ventas):
    lista = []
    for g in generador(ventas):
        lista.append(f"{g['vendedor']} - {g['monto']}")

    return lista

vendedor_monto(ventas)
```

```python
# 4. Crea una función que guarde los resultados 
# del ejercicio anterior en un archivo .txt
# usando context manager. Una venta por línea.
def save():
    with open("ventas.txt", "w") as archivo:
        for linea in vendedor_monto(ventas):
            archivo.write(linea + "\n")

save()
```
