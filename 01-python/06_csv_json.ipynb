```python
import csv

ventas = [
    {"vendedor": "Ana", "mes": "enero", "monto": 15000, "region": "norte"},
    {"vendedor": "Luis", "mes": "enero", "monto": 8000, "region": "sur"},
    {"vendedor": "Ana", "mes": "febrero", "monto": 20000, "region": "norte"},
    {"vendedor": "Eva", "mes": "enero", "monto": 12000, "region": "norte"},
    {"vendedor": "Luis", "mes": "febrero", "monto": 9500, "region": "sur"},
]

with open("ventas.csv", "w", newline="") as archivo:
    campos = ["vendedor", "mes", "monto", "region"]
    writer = csv.DictWriter(archivo, fieldnames=campos)

    writer.writeheader()
    writer.writerows(ventas)

print("Archivo creado")
```

```python
import json

ventas = [
    {"vendedor": "Ana", "mes": "enero", "monto": 15000, "region": "norte"},
    {"vendedor": "Luis", "mes": "enero", "monto": 8000, "region": "sur"},
    {"vendedor": "Ana", "mes": "febrero", "monto": 20000, "region": "norte"},
    {"vendedor": "Eva", "mes": "enero", "monto": 12000, "region": "norte"},
    {"vendedor": "Luis", "mes": "febrero", "monto": 9500, "region": "sur"},
]

# escribir JSON
with open("ventas.json", "w") as archivo:
    json.dump(ventas, archivo, indent=4)

print("Archivo creado")
```

```python
with open("ventas.json", "r") as archivo:
    ventas_json = json.load(archivo)

print(type(ventas_json))
print(type(ventas_json[0]['monto']))
print(ventas_json[0])
```

```python
# 1. Lee el ventas.csv, convierte los montos a int
#    y guarda solo las ventas de región norte en un nuevo archivo
#    llamado "norte.csv"

with open("ventas.csv", "r") as archivo:
    reader = csv.DictReader(archivo)
    filas_leidas = []
    for fila in reader:
        fila['monto'] = int(fila['monto'])
        filas_leidas.append(fila)
        print(fila)
        
    
norte = [f for f in filas_leidas if f['region'] == 'norte']

with open("norte.csv", "w", newline="") as archivo:
    campos = ['vendedor', 'mes', 'monto', 'region']
    writer = csv.DictWriter(archivo, fieldnames=campos)


    writer.writeheader()
    writer.writerows(norte)
```

```python
# 2. Lee el ventas.json y calcula el total de ventas
#    por vendedor — resultado como diccionario

with open("ventas.json", "r") as archivo:
    ventas_json = json.load(archivo)

total = {}
for vj in ventas_json:
    nombre = vj['vendedor']

    if nombre not in total:
        total[nombre] = 0
    
    total[nombre] += vj['monto']

print(total)
```

```python
# 3. Guarda ese diccionario de totales como "totales.json"

with open("totales.json", "w") as archivo:
    json.dump(total, archivo, indent=4)

print("Archivo creado")
```

```python
# 4. Lee "totales.json" e imprime el vendedor con mayor total

with open("totales.json", "r") as archivo:
    totales_leido = json.load(archivo)


mayor = max(totales_leido, key=lambda k: totales_leido[k])
print(mayor)
```
