```python
def dividir(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("No se puede dividir entre cero")
        return None
    finally:
        print("Esto siempre se ejecuta")

print(dividir(10, 0))
print(dividir(10, 2))
```

```python
def procesar(valor):
    try:
        numero = int(valor)
        resultado = 100 / numero
        return resultado
    except ZeroDivisionError:
        print("No puede ser cero")
        return None
    except ValueError:
        print(f"'{valor}' no es un número válido")
        return None
    
print(procesar("abc"))
print(procesar(0))
print(procesar(5))
```

```python
class DatoInvalidoError(Exception):
    pass

def validar_monto(monto):
    if monto < 0:
        raise DatoInvalidoError(f"El monto no puede ser negativo: {monto}")
    return monto

try:
    validar_monto(-500)
except DatoInvalidoError as e:
    print(f"Error de validacion: {e}")
```

```python
# Error personalizado
class DatoInvalidoError(Exception):
    pass

# Clase normal que usa el error
class Dataset:
    def __init__(self, datos):
        if not datos:
            raise DatoInvalidoError("El dataset no puede estar vacío")
        self.datos = datos
    
    def promedio(self):
        return sum(self.datos) / len(self.datos)

# Probando
try:
    ds = Dataset([])
except DatoInvalidoError as e:
    print(f"Error: {e}")

ds = Dataset([10, 20, 30])
print(ds.promedio())
```

```python
ventas = [
    {"vendedor": "Ana", "mes": "enero", "monto": 15000, "region": "norte"},
    {"vendedor": "Luis", "mes": "enero", "monto": 8000, "region": "sur"},
    {"vendedor": "Ana", "mes": "febrero", "monto": 20000, "region": "norte"},
    {"vendedor": "Eva", "mes": "enero", "monto": 12000, "region": "norte"},
    {"vendedor": "Luis", "mes": "febrero", "monto": 9500, "region": "sur"},
]
```

```python
# 1. Crea un error personalizado llamado MontoInvalidoError.
# Crea una función que reciba un monto y lance ese error 
# si el monto es negativo o cero.

class MontoInvalidoError(Exception):
    pass

def recibir_monto(monto):
    if monto  < 0 or monto == 0:
        raise MontoInvalidoError("El monto no puede ser negativo o cero")
    
    print("Monto aceptado")
    
try:
    recibir_monto(-ventas[2]["monto"])
except MontoInvalidoError as e:
    print(f"Error: {e}")
```

```python
# 2. Crea una función que reciba la lista de ventas y un vendedor.
# Si el vendedor no existe en la lista, lanza un error personalizado
# llamado VendedorNoEncontradoError.
# Si existe, retorna el total de sus ventas.

class VendedorNoEncontradoError(Exception):
    pass

def total_ventas(ventas: list, vendedor):

    nombre = [v for v in ventas if v['vendedor'] == vendedor]

    if not nombre:
        raise VendedorNoEncontradoError("Ese vendedor no existe")
    
    total = sum(v['monto'] for v in nombre)
    return total

try:
    print(total_ventas(ventas, "Ana"))
except VendedorNoEncontradoError as e:
    print(f"Error: {e}")
```

```python
# 3. Crea una función que guarde la lista de ventas en un archivo .txt
# Si algo falla al escribir, atrapa el error e imprime el mensaje.
# Usa finally para imprimir "Operación terminada" siempre.

def escribir(ventas: list):
    try:
        with open("info.txt", "w") as archivo:
            for linea in ventas:
                archivo.write(str(linea) + "\n")
    except Exception as e:
        print(f"Error: {e}")
    finally:
        print("Operacion terminada")

escribir(ventas)
```
