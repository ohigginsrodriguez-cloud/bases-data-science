```python
def dividir(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("No se puede dividir entre cero")
        return None
    finally:
        print("Esto siempre se ejecuta")

print(dividir(10, 0))
print(dividir(10, 2))
```

```python
def procesar(valor):
    try:
        numero = int(valor)
        resultado = 100 / numero
        return resultado
    except ZeroDivisionError:
        print("No puede ser cero")
        return None
    except ValueError:
        print(f"'{valor}' no es un número válido")
        return None
    
print(procesar("abc"))
print(procesar(0))
print(procesar(5))
```

```python
class DatoInvalidoError(Exception):
    pass

def validar_monto(monto):
    if monto < 0:
        raise DatoInvalidoError(f"El monto no puede ser negativo: {monto}")
    return monto

try:
    validar_monto(-500)
except DatoInvalidoError as e:
    print(f"Error de validacion: {e}")
```

```python
# Error personalizado
class DatoInvalidoError(Exception):
    pass

# Clase normal que usa el error
class Dataset:
    def __init__(self, datos):
        if not datos:
            raise DatoInvalidoError("El dataset no puede estar vacío")
        self.datos = datos
    
    def promedio(self):
        return sum(self.datos) / len(self.datos)

# Probando
try:
    ds = Dataset([])
except DatoInvalidoError as e:
    print(f"Error: {e}")

ds = Dataset([10, 20, 30])
print(ds.promedio())
```
