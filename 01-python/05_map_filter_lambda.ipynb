```python
# funcion normal
def doble(x):
    return x * 2

# lambda - funcion en una linea
doble_lambda = lambda x: x * 2

print(doble(5))
print(doble_lambda(5))
```

```python
numeros = [1, 2, 3, 4, 5]

# sin map
dobles = []
for n in numeros:
    dobles.append(n * 2)
    
# con map
dobles = list(map(lambda x: x * 2, numeros))

print(dobles)
```

```python
numeros = [1, 2, 3, 4, 5, 6]

# solo los pares
pares = list(filter(lambda x: x % 2 == 0, numeros))
print(pares)

pares_comprehesion = [n for n in numeros if n % 2 == 0]
print(pares_comprehesion)
```

```python
ventas = [
    {"vendedor": "Ana", "monto": 15000},
    {"vendedor": "Luis", "monto": 8000},
    {"vendedor": "Eva", "monto": 12000},
]

# ordenar por monto
ordenadas = sorted(ventas, key=lambda v: v['monto'])
print(ordenadas)

# de mayor a menor
ordenadas_desc = sorted(ventas, key=lambda v: v['monto'], reverse=True)
print(ordenadas_desc)
```
