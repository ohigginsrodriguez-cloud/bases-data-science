```python
# funcion normal
def doble(x):
    return x * 2


# lambda - funcion en una linea
doble_lambda = lambda x: x * 2

print(doble(5))
print(doble_lambda(5))
```

```python
numeros = [1, 2, 3, 4, 5]

# sin map
dobles = []
for n in numeros:
    dobles.append(n * 2)

# con map
dobles = list(map(lambda x: x * 2, numeros))

print(dobles)
```

```python
numeros = [1, 2, 3, 4, 5, 6]

# solo los pares
pares = list(filter(lambda x: x % 2 == 0, numeros))
print(pares)

pares_comprehesion = [n for n in numeros if n % 2 == 0]
print(pares_comprehesion)
```

```python
ventas = [
    {"vendedor": "Ana", "monto": 15000},
    {"vendedor": "Luis", "monto": 8000},
    {"vendedor": "Eva", "monto": 12000},
]

# ordenar por monto
ordenadas = sorted(ventas, key=lambda v: v["monto"])
print(ordenadas)

# de mayor a menor
ordenadas_desc = sorted(ventas, key=lambda v: v["monto"], reverse=True)
print(ordenadas_desc)
```

```python
ventas = [
    {"vendedor": "Ana", "mes": "enero", "monto": 15000, "region": "norte"},
    {"vendedor": "Luis", "mes": "enero", "monto": 8000, "region": "sur"},
    {"vendedor": "Ana", "mes": "febrero", "monto": 20000, "region": "norte"},
    {"vendedor": "Eva", "mes": "enero", "monto": 12000, "region": "norte"},
    {"vendedor": "Luis", "mes": "febrero", "monto": 9500, "region": "sur"},
]
```

```python
# 1. Usando map, crea una lista con los montos de todas las ventas
montos = list(map(lambda v: v["monto"], ventas))
print(montos)
```

```python
# 2. Usando filter, obtén solo las ventas del mes de enero
enero = list(filter(lambda v: v["mes"] == "enero", ventas))
for e in enero:
    print(e)
```

```python
# 3. Usando sorted, ordena las ventas por vendedor alfabéticamente
alfabeto = sorted(ventas, key=lambda v: v["vendedor"])
for a in alfabeto:
    print(a)
```

```python
# 4. Combina filter y sorted: obtén las ventas de la región norte
#    ordenadas de mayor a menor monto
sort_nort = sorted(
    filter(lambda v: v["region"] == "norte", ventas),
    key=lambda v: v["monto"],
    reverse=True,
)

for sn in sort_nort:
    print(sn)
```

```python
# 5. Crea una función llamada aplicar que reciba la lista de ventas
#    y una función, y retorne el resultado de aplicar esa función
#    a la lista. Pruébala con una lambda que extraiga los montos.
def aplicar(ventas: list, funcion):
    return list(map(funcion, ventas))


funcion = lambda v: v["monto"]

print(aplicar(ventas, funcion))
```
