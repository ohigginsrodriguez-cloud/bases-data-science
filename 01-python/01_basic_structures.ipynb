```python
ventas = [
    {"vendedor": "Ana", "mes": "enero", "monto": 15000, "region": "norte"},
    {"vendedor": "Luis", "mes": "enero", "monto": 8000, "region": "sur"},
    {"vendedor": "Ana", "mes": "febrero", "monto": 20000, "region": "norte"},
    {"vendedor": "Eva", "mes": "enero", "monto": 12000, "region": "norte"},
    {"vendedor": "Luis", "mes": "febrero", "monto": 9500, "region": "sur"},
]
```

```python
# Solo region por elemento
for v in ventas:
    print(v['region'])
```

```python
# Vendedor y region juntos, misma linea
for v in ventas:
    print(f"{v['vendedor']} - {v['region']}")
```

```python
# Solo vendedores region norte
for v in ventas:
    if v['region'] == "norte":
        print(v['vendedor'])
```

```python
# List comprehesion de solo vendedores region norte
vendedores_norte = [v['vendedor'] for v in ventas if v['region'] == 'norte']
print(vendedores_norte)
```

```python
totales = {}

for v in ventas:
    nombre = v['vendedor']

    if nombre not in totales:
        totales[nombre] = 0

    totales[nombre] += v['monto']

    print(totales)
```

```python
# Contar cuantas ventas hizo cada vendedor
conteo_ventas = {}

for v in ventas:
    nombre = v['vendedor']

    if nombre not in conteo_ventas:
        conteo_ventas[nombre] = 0

    conteo_ventas[nombre] += 1

print(conteo_ventas)
```

```python
for v in ventas:
    par = (v['vendedor'], v['monto'])
    print(par)
```

```python
pares = [(v['vendedor'], v['monto']) for v in ventas]
print(pares)
```

```python
# 1. Lista de vendedores únicos

nombres = [v['vendedor'] for v in ventas]
unicos = set(nombres)
print(unicos)
```

```python
# 2. Total de ventas por vendedor
total = {}
for v in ventas:
    nombre = v['vendedor']
    if nombre not in total:
        total[nombre] = 0

    total[nombre] += v['monto']

print(total)
```

```python
# 3. Ventas de región norte mayores a 10000
mayores_norte = [v for v in ventas if v['region'] == 'norte' and v['monto'] > 10000]
print(mayores_norte)
```

```python
# 4. Función que reciba la lista y un vendedor, retorne su promedio
def average(ventas: list, vendedor_nombre):
    count = 0
    total = 0
    for v in ventas:
        if v['vendedor'] == vendedor_nombre:
            total += v['monto']
            count += 1
            
    promedio = total / count

    if count == 0:
        return None

    return promedio

average(ventas, "Ana")


        
```

```python
# 5. Lista de tuplas (vendedor, monto)
pares = [(v['vendedor'], v['monto']) for v in ventas]
print(pares)
```
