```python
class Vendedor:
    def __init__(self, nombre, region):
        self.nombre = nombre
        self.region = region
        self.ventas = []

    def agregar_venta(self, monto):
        self.ventas.append(monto)

    def total(self):
        return sum(self.ventas)

    # para escribir el objeto de forma legible al imprimir
    def __repr__(self):
        return f"Vendedor({self.nombre}, {self.region})"

    def __len__(self):
        return len(self.ventas)

    def __str__(self):
        return f"Vendedor: {self.nombre}, Region: {self.region}, Ventas: {self.ventas}"


ana = Vendedor("Ana", "norte")
ana.agregar_venta(15000)
ana.agregar_venta(20000)

print(ana)
print(ana.total())
print(len(ana))
print(str(ana))
```

```python
class Empleado:
    def __init__(self, nombre, region):
        self.nombre = nombre
        self.region = region

    def presentarse(self):
        return f"Soy: {self.nombre} de {self.region}"


class Vendedor(Empleado):
    def __init__(self, nombre, region):
        super().__init__(nombre, region)
        self.ventas = []

    def agregar_venta(self, monto):
        self.ventas.append(monto)


ana = Vendedor("Ana", "norte")

print(ana.presentarse())  # metodo heredado de empleado
ana.agregar_venta(15000)
print(ana.ventas)
```

```python
class ModeloSimple:
    def __init__(self):
        self.promedio = None
        self.entrenado = False

    def fit(self, datos):
        self.promedio = sum(datos) / len(datos)
        self.entrenado = True
        return self

    def predict(self, datos):
        if not self.entrenado:
            raise Exception("Primero debes entrenar el modelo con .fit()")
        return [self.promedio for _ in datos]


modelo = ModeloSimple()
modelo.fit([10, 20, 30, 40])
print(modelo.predict([1, 2, 3]))
```

```python
# 1. Crea una clase Vendedor con:
#    - __init__ que reciba nombre y region
#    - atributo ventas como lista vacía
#    - método agregar_venta(monto) que valide que el monto sea positivo
#      y lance un MontoInvalidoError si no lo es
#    - método total() que retorne la suma de sus ventas
#    - método promedio() que retorne el promedio de sus ventas
#    - __str__ que muestre nombre, region y total


class MontoInvalidoError(Exception):
    pass


class VentasVaciasError(Exception):
    pass


class Vendedor:
    def __init__(self, nombre, region):
        self.nombre = nombre
        self.region = region
        self.ventas = []

    def agregar_venta(self, monto):
        if monto < 0:
            raise MontoInvalidoError("El monto tiene que ser postivo")

        self.ventas.append(monto)

    def total(self):
        if not self.ventas:
            raise VentasVaciasError("Debes tener ventas, esta vacia la lista")

        return sum(self.ventas)

    def promedio(self):
        if not self.ventas:
            raise VentasVaciasError(
                "Para promediar hay que tener elementos en la lista"
            )

        return sum(self.ventas) / len(self.ventas)

    def __str__(self):
        return f"Nombre: {self.nombre}, Region: {self.region}, Total: {self.total()}"


try:
    ana = Vendedor("Ana", "norte")
    ana.total()
    ana.promedio()
    ana.agregar_venta(100)
    ana.agregar_venta(300)
    print(ana.total())
    print(ana.promedio())
    print(ana)
except MontoInvalidoError as e:
    print(f"Error: {e}")
except VentasVaciasError as e:
    print(f"Error: {e}")
finally:
    print("Proceso terminado")
```

```python
# 2. Crea una clase Dataset con:
#    - __init__ que reciba la lista de ventas (los dicts de arriba)
#    - método cargar() que cree un objeto Vendedor por cada vendedor único
#      y les agregue sus ventas correspondientes
#    - método resumen() que imprima nombre, total y promedio de cada vendedor


class Dataset:
    def __init__(self, ventas: list):
        self.ventas = ventas
        self.vendedores = {}

    def cargar(self, ventas):
        for v in self.ventas:
            nombre = v["vendedor"]

            if nombre not in self.vendedores:
                self.vendedores[nombre] = Vendedor(v["vendedor"], v["region"])

            self.vendedores[nombre].agregar_venta(v["monto"])

    def resumen(self):
        for v in self.vendedores.values():
            print(f"Nombre: {v.nombre}, Total: {v.total()}, Promedio: {v.promedio()}")
```

```python
# 3. Usa Dataset para cargar los datos e imprime el resumen
data = Dataset(ventas)
data.cargar(ventas)
data.resumen()
```
