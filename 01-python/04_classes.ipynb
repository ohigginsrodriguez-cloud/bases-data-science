```python
class Vendedor:
    def __init__(self, nombre, region):
        self.nombre = nombre
        self.region = region
        self.ventas = []

    def agregar_venta(self, monto):
        self.ventas.append(monto)

    def total(self):
        return sum(self.ventas)
    
    def __repr__(self):
        return f"Vendedor({self.nombre}, {self.region})"
    
ana = Vendedor("Ana", "norte")
ana.agregar_venta(15000)
ana.agregar_venta(20000)

print(ana)
print(ana.total())
```
