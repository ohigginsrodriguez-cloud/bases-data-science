```python
import numpy as np
from scipy import stats
```

```python
# Ventas del norte vs sur
norte = [15000, 20000, 12000, 18000, 15000, 9000, 22000]
sur = [8000, 9500, 11000]

t_stat, p_value = stats.ttest_ind(norte, sur)

print(f"p-value: {round(p_value, 4)}")

if p_value < 0.05:
    print("La diferencia es estadísticamente significativa")
else:
    print("La diferencia podría ser aleatoria")
```

```python
# Grupos muy similares
norte2 = [15000, 16000, 14000, 15500, 14500]
sur2 = [14500, 15500, 15000, 14800, 15200]

t_stat, p_value = stats.ttest_ind(norte2, sur2)
print(f"p-value grupos similares: {round(p_value, 4)}")
```
