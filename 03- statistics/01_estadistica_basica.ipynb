```python
import numpy as np
from scipy.stats import mode
```

```python
datos = [15000, 8000, 20000, 12000, 9500, 18000, 15000, 11000, 9000, 22000]

print("Media:", np.mean(datos))
print("Mediana:", np.median(datos))
print("Moda:", mode(datos).mode)
print("Desviación estándar:", np.std(datos))
print("Varianza:", np.var(datos))
print("Mínimo:", np.min(datos))
print("Máximo:", np.max(datos))
```

```python
print("Diferencia media-mediana:", np.mean(datos) - np.median(datos))
print("Rango: ", np.max(datos) - np.min(datos))
```

```python
print("Percentil 25:", np.percentile(datos, 25))
print("Percentil 50:", np.percentile(datos, 50))
print("Percentil 75:", np.percentile(datos, 75))
print("IQR:", np.percentile(datos, 75) - np.percentile(datos, 25))
```

```python
q25 = np.percentile(datos, 25)
q75 = np.percentile(datos, 75)
iqr = q75 - q25

limite_bajo = q25 - 1.5 * iqr
limite_alto = q75 + 1.5 * iqr

print("Límite bajo:", limite_bajo)
print("Límite alto:", limite_alto)
print("Outliers:", [x for x in datos if x < limite_bajo or x > limite_alto])
```

```python
import numpy as np

ventas = [15000, 8000, 20000, 12000, 9500, 18000, 15000, 11000, 9000, 22000]
experiencia = [5, 1, 7, 3, 2, 6, 4, 3, 2, 8]  # años de experiencia

correlacion = np.corrcoef(ventas, experiencia)[0, 1]
print("Correlacion:", round(correlacion, 4))
```

```python
sin_relacion = [3, 8, 1, 9, 2, 7, 4, 6, 5, 10]

correlacion2 = np.corrcoef(ventas, sin_relacion)[0, 1]
print("Correlacion sin relacion:", round(correlacion2, 4))
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

#Con correlacion alta
axes[0].scatter(experiencia, ventas)
axes[0].set_title(f"Correlacion alta: 0.9919")
axes[0].set_xlabel("Experiencia")
axes[0].set_ylabel("ventas")

#Sin correlacion
axes[1].scatter(sin_relacion, ventas)
axes[1].set_title(f"Sin correlacion: 0.0435")
axes[1].set_xlabel("Variable aleatoria")
axes[1].set_ylabel("Ventas")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Distribución normal — la más importante en estadística
datos_normales = np.random.normal(loc=50, scale=10, size=1000)

plt.figure(figsize=(10, 4))
plt.hist(datos_normales, bins=30, edgecolor='black')
plt.title("Distribución normal")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()

print("Media:", round(np.mean(datos_normales), 2))
print("Desviación estándar:", round(np.std(datos_normales), 2))
```

```python
media = np.mean(datos_normales)
std = np.std(datos_normales)

dentro_1std = np.sum((datos_normales >= media - std) & (datos_normales <= media + std))
dentro_2std = np.sum((datos_normales >= media - 2*std) & (datos_normales <= media + 2*std))
dentro_3std = np.sum((datos_normales >= media - 3*std) & (datos_normales <= media + 3*std))

print(f"Dentro de 1 std: {dentro_1std/10}%")
print(f"Dentro de 2 std: {dentro_2std/10}%")
print(f"Dentro de 3 std: {dentro_3std/10}%")
```

```python
# Distribución sesgada — como salarios o precios de casas
datos_sesgados = np.random.exponential(scale=10, size=1000)

# Distribución uniforme — todos los valores igual de probables
datos_uniformes = np.random.uniform(low=0, high=100, size=1000)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(datos_sesgados, bins=30, edgecolor='black')
axes[0].set_title("Distribución sesgada (exponencial)")

axes[1].hist(datos_uniformes, bins=30, edgecolor='black')
axes[1].set_title("Distribución uniforme")

plt.tight_layout()
plt.show()

print("Media:", round(np.mean(datos_sesgados), 2))
print("Mediana:", round(np.median(datos_sesgados), 2))
```
